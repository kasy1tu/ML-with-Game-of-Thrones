# ATTACKERS

```python
# List of kings who initiated attacks in GoT
attacker_kings = list(battles_df["attacker_king"].value_counts().index)
attacker_kings
```

```python
# Value counts: number of attacks per king
battles_df["attacker_king"].value_counts()
```

```python
# Integer values of attacks per king
attacks_list = list(battles_df["attacker_king"].value_counts().values)
attacks_list

int_attacks = []

for attack in attacks_list:
    int_attacks.append(int(attack))
    
int_attacks
attacks = int_attacks
attacks
```

# DEFENDERS

```python
# List of kings who defended in GoT
defender_kings = list(battles_df["defender_king"].value_counts().index)
defender_kings
```

```python
# Value Counts: number of defenses per king
battles_df["defender_king"].value_counts()
```

```python
# Integer values of defenses per king
defenses_list = list(battles_df["defender_king"].value_counts().values)
defenses_list

int_defenses = []

for defense in defenses_list:
    int_defenses.append(int(defense))
    
defenses = int_defenses
defenses
```

# Making a new data frame

```python
# List of kings
kings = []

for king in defender_kings:
    kings.append(king)
    
kings

for king in attacker_kings:
    if king not in kings:
        kings.append(king)

kings
```

```python
# Cleaned DF of just kings, number of attacks, number of defenses

data = {
    "King": kings,
    "Attacks": [10, 14, 7, 5, 0, 0],
    "Defenses": [14, 13, 4, 2, 1, 1]
}

new_df = pd.DataFrame(data)
new_df.to_csv("kings.csv", index = False)
new_df
```
