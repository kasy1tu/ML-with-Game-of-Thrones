```python
import pandas as pd
```

```python
# Load in csv
script_df = pd.read_csv("../data/scripts.csv")
script_df.head()
```

```python

filtered = script_df[ script_df['Name']=='jon snow' ]
filtered.head()
```

```python
lines = filtered['Sentence'].to_numpy()
linestring = ""
for line in lines:
    linestring = linestring + str(line) + " "
print(linestring)
```

```python

print(linestring, file=open("jon_snow.txt", "w"))
```
